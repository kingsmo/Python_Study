## Everytime Crawling Test

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from time import sleep
```

```python
# Using Chrome Webdriver

driver = webdriver.Chrome('/Users/sungjae/Downloads/chromedriver')
```

```python
# Login into Everytime

driver.get('https://everytime.kr/login')

driver.find_element_by_name('userid').send_keys('usrid')
driver.find_element_by_name('password').send_keys('usrpass')

driver.find_element_by_class_name('submit').click()
```

```python
base_url = 'https://everytime.kr/374614/p/'
url = base_url + str(1)
print(url)
```

```python
# go to kookmin university pages

base_url = 'https://everytime.kr/374614/p/'

for i in range(1, 2):
    all_page = base_url + str(i)
    driver.get(all_page)
    
    sleep(1)
    
    page_html = driver.page_source
    soup = BeautifulSoup(page_html, 'html.parser')
    articles = soup.select('#container > div.articles > article > a.article')
    
    if len(articles) < 1:
        print('error')
    
    for a in articles:
        single_link = 'https://everytime.kr' + str(a['href'].strip())
        
        driver.get(single_link)
         
        sleep(0.5)
        
        single_page_html = driver.page_source
        soup = BeautifulSoup(single_page_html, 'html.parser')
        single_article = soup.select('#container > div.articles > article > a.article')
        
        if len(single_article) < 1:
            print('single error')
        
        for s in single_article:
            article_title = s.select('h2.large')[0].text.strip()
            article_content = s.select('p.large')[0].text.strip()
            article_time = s.select('time.large')[0]['title']
            article_like = s.select('li.vote')[0].text.strip()
            article_comment = s.select('li.comment')[0].text.strip()
            article_scrap = s.select('li.scrap')[0].text.strip()
            print('제목 :', article_title)
            print('본문 :', article_content)
            print('시간 :', article_time)
            print('좋아요 :', article_like)
            print('댓글 :', article_comment)
            print('스크랩 :', article_scrap)
            print()
```
