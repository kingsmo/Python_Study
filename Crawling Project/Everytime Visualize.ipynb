## Everytime Visualize

```python
import pandas as pd
import sqlite3
```

```python
sql_query = 'SELECT * FROM Contents'
conn = sqlite3.connect('everytime_db.sqlite')
```

```python
everytime = pd.read_sql(sql_query, con=conn)
```

```python
everytime[everytime['like']==everytime['like'].max()]
```

```python
everytime[everytime['comments']==everytime['comments'].max()]
```

```python
everytime['time']
```

```python
everytime.time[0][8:10]
```

```python
everytime['day'] = everytime.time.apply(lambda x: x[8:10])
# pre1['happenYear'] = pre1.happenStr.apply(lambda x: x[:4])
```

```python
everytime['day']
```

```python
everytime19 = everytime[everytime['day'] == '19']
```

```python
everytime19.count()
```

```python
everytime19['time'][100][11:13]
```

```python
everytime19['hour'] = everytime19.time.apply(lambda x: x[11:13])
everytime19['hour']
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
everytime19_count = everytime19.groupby(['hour'])['hour'].count()
```

```python
everytime19_count
```

```python
plt.rcParams["figure.figsize"] = (12,4)
plt.rcParams['lines.linewidth'] = 2
plt.plot(everytime19_count, color = 'b');
```
